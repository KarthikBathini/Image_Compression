```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
from sklearn.cluster import KMeans
```

```python
def compress(path,k=16):
    image=cv2.imread(path)
    image=cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    pixels = image.reshape(-1,3)
    kmeans = KMeans(n_clusters=k, random_state =42, n_init=10)
    kmeans.fit(pixels)
    compressed_pixels= kmeans.cluster_centers_[kmeans.labels_]
    compressed_image = compressed_pixels.reshape(image.shape).astype(np.uint8)
    return image, compressed_image
```

```python
def display_image(uploaded,compressed):
    plt.imshow(uploaded.astype(np.uint8))
    plt.title('Uploaded Image')
    plt.axis('off')
    plt.show()

    plt.imshow(compressed.astype(np.uint8))
    plt.title("Compressed Image")
    plt.axis('off')
    plt.show()
    
```

```python
image_path= "C:/Users/nbath/OneDrive/Pictures/Hero 1.jpg";
uploaded, compressed=compress(image_path,k=16)
display_image(uploaded,compressed)
```
